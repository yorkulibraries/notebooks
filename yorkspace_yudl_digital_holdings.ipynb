```python
import xml.etree.ElementTree as ET
import pandas as pd
import glob
import os
```

```python
cols = ["file",
        "title",
        "field_alternative_title",
        "created",
        "field_local_identifier",
        "field_hdl",
        "field_language",
        "field_genre",
        "field_date",
        "field_extent",
        "field_publisher",
        "field_linked_agent",
        "field_subject",
        "field_member_of",
        "field_rights",
        "field_description"]
rows = []
```

```python
data_path = "/content/1461-dc"
filelist = glob.glob(data_path + "/*.xml")
```

```python
for f in filelist:
    tree = ET.parse(f)
    root = tree.getroot()

    try:
      rights = root.find(".//dcvalue[@element='rights'][@qualifier='none']").text
    except:
      rights = ""
    try:
      alternative_title = root.find(".//dcvalue[@element='title'][@qualifier='alternative']").text
    except:
      alternative_title = ""
    try:
      member_of = '|'.join([member.text for membership in root.findall(".//dcvalue[@element='relation'][@qualifier='ispartofseries']")])
    except:
      member_of = ""
    try:
      hdl = root.findall(".//dcvalue[@element='identifier'][@qualifier='uri']")[0].text
    except:
      hdl = ""
    try:
      identifier = root.find(".//dcvalue[@element='identifier'][@qualifier='none']").text
    except:
      identifier = ""
    try:
      filename = root.find(".//dcvalue[@element='identifier'][@qualifier='none']").text + ".pdf"
    except:
      filename = ""
    title = root.find(".//dcvalue[@element='title'][@qualifier='none']").text
    try:
      genre = root.find(".//dcvalue[@element='type'][@qualifier='none']").text
    except:
      genre = ""
    try:
      language = root.find(".//dcvalue[@element='language'][@qualifier='iso']").text
    except:
      language: ""
    try:
      extent = root.find(".//dcvalue[@element='format'][@qualifier='none']").text
    except:
      extent = ""
    try:
      date = root.find(".//dcvalue[@element='date'][@qualifier='issued']").text
    except:
      date = ""
    created = root.find(".//dcvalue[@element='date'][@qualifier='accessioned']").text
    try:
      publisher = root.find(".//dcvalue[@element='publisher'][@qualifier='none']").text
    except:
      publisher = ""
    try:
      creator = '|'.join([creators.text for creators in root.findall(".//dcvalue[@element='creator'][@qualifier='author']")])
    except:
      creator = ""
    subject = '|'.join([subjects.text for subjects in root.findall(".//dcvalue[@element='subject'][@qualifier='none']")])
    description = '\n'.join(filter(None, [descriptions.text for descriptions in root.findall(".//dcvalue[@element='description']")]))

    rows.append({"file": filename,
            "title": title,
            "field_alternative_title": alternative_title,
            "created": created,
            "field_local_identifier": identifier,
            "field_hdl": hdl,
            "field_language": language,
            "field_genre": genre,
            "field_date": date,
            "field_extent": extent,
            "field_publisher": publisher,
            "field_linked_agent": creator,
            "field_subject": subject,
            "field_member_of": member_of,
            "field_rights": rights,
            "field_description": description})
```

```python
df = ""
```

```python
df = pd.DataFrame(rows, columns=cols)
```
